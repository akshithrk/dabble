```python
raw_data = pd.read_table('yesware-test-data-v1-6.txt', names=('A'))
```

```python
df = pd.DataFrame(raw_data)
df.columns=['original']
```

```python
# below line parses the entire df, removes everything after the '--' & writes data to a new column called split
df['split'] = df['original'].map(lambda x: x.split('--', 1)[0])
```

```python
# below line displays only every 2nd row in the df
df.iloc[::2]
```

```python
# below to show if the data in the column contains the specified character
# useing this column in the future to check if any non name rows get selected in new_df
df['required'] = df.split.str.contains(',') == True
```

```python
# skip alternate row using indexs
new_df = df.iloc[::2]
```

```python
df
```

```python
# rename column split to column full_name 
new_df = new_df.rename(columns = {'split':'full_name'})
```

```python
new_df
```

```python
# parse full_name into first and last
new_df['last'] = new_df['full_name'].map(lambda x: x.split(',', 1)[0])
new_df['first'] = new_df['full_name'].map(lambda x: x.split(',', -1)[-1])
# removing the test column from new_df
del new_df['required']
```

```python
del new_df['original']
```

```python
# noticed that there was a whitespace at the end to stripping it
new_df['full_name'] = new_df['full_name'].str.rstrip()
```
